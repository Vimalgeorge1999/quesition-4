```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Lenovo\Downloads\train.csv")
df
```

```python
#The data is too big and we only need to consider the important columns,
```

```python
df=df[["LotFrontage","LotArea","Street","RoofStyle","HouseStyle","BedroomAbvGr","KitchenAbvGr","GarageArea","SalePrice"]]
df
```

# data preprocessing

```python
df.isnull().sum()/len(df)*100
```

```python
#fixing missing vals
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='mean')
si
```

```python
df["LotFrontage"]=si.fit_transform(df[["LotFrontage"]])
df
```

```python
df.isnull().sum()   # all missing values are fixed
```

# feature encoding

```python
df_ohe=pd.get_dummies(df[["Street","RoofStyle","HouseStyle"]])
df_ohe
```

```python
df=pd.concat([df,df_ohe],axis=1)
df
```

```python
df=df.drop(["Street","RoofStyle","HouseStyle"],axis=1)
df
```

# feature scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss
```

```python
df[["LotFrontage","LotArea","GarageArea"]]=ss.fit_transform(df[["LotFrontage","LotArea","GarageArea"]])
df
```

# DIVIDE X AND Y

```python
x=df.drop(["SalePrice"],axis=1)
y=df["SalePrice"]
```

# split train test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

# fit and apply linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```
